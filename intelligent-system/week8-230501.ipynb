# seabon

- 독립관계 : 아무런 관련이 없음
- 상관관계 : 두 개의 변수들이 함께 변화하는 관계
- 인과관계 : 한 변수가 다른 변수의 직접적인 원인이 됨
- 상호관계 : 두 변수가 서로 영향을 주는 관계

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

상관계수
corrcoed()

```python
np.random.seed(85)
x= np.arange(0,10)
y1=x*2
np.corrcoef(x,y1)
```

```python
# x=np.arange(0,10)
y2=x**3
```

```python
np.corrcoef(x,y2)
```

```python
np.random.seed(85)
y3=np.random.randint(0,100,size=10)
np.corrcoef(x,y3)
```

```python
result=np.corrcoef((x,y2,y3)) #튜플
print(result)
```

```python
plt.imshow(result)
plt.colorbar()
```

```python
df =pd.DataFrame({'x': x, "y2":y2, "y3":y3})
sns.pairplot(df)
```
